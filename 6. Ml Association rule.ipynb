```python
import pandas as pd
import csv
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules # Corrected the typo in the function name from 'asssociation_rules' to 'association_rules'
```

```python
dataset=[]
import csv
with open('Market_Basket_Optimisation.csv') as file:
    reader=csv.reader(file,delimiter=',')
    for row in reader:
      dataset += [row]

```

```python
te = TransactionEncoder()
x= te.fit_transform(dataset)
```

```python
x
```

```python
df = pd.DataFrame(x, columns=te.columns_)  # Use te.columns_ instead of te
```

```python
te.columns_
```

```python
 len(te.columns_)
```

```python
df.head()
```

```python
#.find frequent itemsets
freq_itemsets = apriori(df, min_support=0.01,use_colnames=True)
```

```python
freq_itemsets
```

```python
rules=association_rules(freq_itemsets,metric='confidence',min_threshold=0.25)
```

```python
rules = rules[['antecedents', 'consequents', 'support', 'confidence']]
```
