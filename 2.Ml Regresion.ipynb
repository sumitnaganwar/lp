```python
import pandas as pd
import matplotlib.pyplot as plt //visualzation
import seaborn as sns  //visualization
```

```python
df=pd.read_csv('temperatures.csv')
```

```python
df
```

```python
df.head()
```

```python
#input data
x=df['YEAR']

#output
y=df['ANNUAL']
```

```python
plt.figure(figsize=(16,9))
plt.title('Temperature plot of INDIA')
plt.xlabel(('Year'))
plt.ylabel('Annual Average Temperature')
plt.scatter(x,y)
```

```python
x.shape
```

```python
x=x.values  #value= convert into array
```

```python
x
```

```python
x=x.reshape(117,1)
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x,y)  #model is train
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
regressor.predict([[2025]])
```

```python
#2
predicted=regressor.predict(x)
```

```python
import numpy as np  # It provides support for large, multi-dimensional arrays and matrices, along with a wide range of mathematical functions
```

```python
# Mean Absolute Error
np.mean(abs(y-predicted))
```

```python
#ALSO we can used inbuit method
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y,predicted)
```

```python
# mean square error
from sklearn.metrics import mean_squared_error
mean_squared_error(y,predicted)
```

```python
#R square matrix
from sklearn.metrics import r2_score
r2_score(y,predicted)
```

```python
regressor.score(x,y)
```

```python
plt.title('Temperature plot of INDIA')
plt.xlabel(('Year'))
plt.ylabel('Annual Average Temperature')
plt.scatter(x,y,label='actual',color='r')
plt.plot(x,predicted,label='predicted',color='g')
plt.legend
```
